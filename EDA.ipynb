```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
telco = pd.read_csv('Churn.csv')
```

```python
telco.head()
```

```python
def dataoveriew(df, message):
    print(f'{message}:\n')
    print("Nombre de lignes:", df.shape[0])
    print("\n Nombre de colonnes :", df.shape[1])
    print("\nFeatures:")
    print(df.columns.tolist())
    print("\nUnique values:")
    print(df.nunique())
dataoveriew(telco, "Aperçu de l'ensemble de données" )
```

```python
pie, ax = plt.subplots(figsize=[10,6])
labels = ['non_Churner', 'Churner']
plt.pie(x=telco['Churn'].value_counts(), autopct="%.1f%%", explode=[0.05]*2, labels=labels, pctdistance=0.5)
plt.title("Taux de Churn", fontsize=14);
#pie.savefig("Taux de Churn.png")
```

Remarque : L'ensemble des donées est déquilibrés (highly unbalenced data)

```python
columns_obj = [col for col in telco.columns if telco[col].dtypes =='object'and telco[col].name!='Phone' and telco[col].name!='Churn']
columns_charge = ['Day_Charge','Eve_Charge','Night_Charge','Intl_Charge',]
```

```python
columns_obj
```

```python
plt.figure(figsize = (12,6))
sns.countplot(data = telco, x = 'Intl_Plan', hue='Churn')
plt.title('Impact de International Plan sur Churner', fontsize = 20)
plt.show()
```

```python
plt.figure(figsize = (12,6))
sns.countplot(data = telco, x = 'Vmail_Plan', hue='Churn')
plt.title('Impact de Voice Mail Plan sur Churner', fontsize = 20)
plt.show()
```

On remarque que les client qui optent pour Voice Mail Plan en génerale ne quittent pas

```python
columns_charge
```

```python
for col in columns_charge:
    sns.boxplot(x='Churn',y=col ,data=telco) 
    plt.show()
```

On remarque que la moyenne des Day_Charge pour churner est plus élevés, ce qui fait penser que peut être les Churner quittent pour des meilleurs prix.

```python
cols = ['Account_Length', 'Vmail_Message','Day_Mins','Day_Calls','Eve_Mins','Eve_Calls','Night_Mins','Night_Calls','Intl_Mins','Intl_Calls']
for col in cols:
    sns.histplot(x=col, data=telco)
    plt.show()
    
```

```python
plt.figure(figsize = (18,15))
plt.subplot(1,3,1)
sns.boxplot(data=telco, y = 'Day_Calls',x='Churn', color = 'lightblue')


plt.subplot(1,3,2)
sns.boxplot(data = telco, y = 'Eve_Calls',x='Churn', color = 'lightgreen')


plt.subplot(1,3,3)
sns.boxplot(data=telco, y = 'Night_Calls', x='Churn',color = 'white')


plt.show()
```

On remarque que la pluspart des appels ont été éffectuer pendant le Matin
Comme prévu Day_chages est elevé pour les Churner  qui ont également un total de Day_calls élevé

```python
plt.figure(figsize = (18,15))
plt.subplot(1,3,1)
sns.boxplot(data=telco, y = 'Day_Mins',x='Churn', color = 'lightblue')


plt.subplot(1,3,2)
sns.boxplot(data = telco, y = 'Eve_Mins',x='Churn', color = 'lightgreen')


plt.subplot(1,3,3)
sns.boxplot(data=telco, y = 'Night_Mins',x='Churn', color = 'white')


plt.show()
```

Malgré le plus grand nombre d'appels passés pendant la journée, nous pouvons voir que les clients ne passent pas plus de temps sur les appels le matin
Churner passent un temps notable des appels le matin

```python
plt.figure(figsize = (15,8))
plt.subplot(1,3,1)
sns.scatterplot(data = telco, x = 'Day_Charge', y = 'Day_Mins',
                hue = 'Churn', alpha = 0.6)

plt.subplot(1,3,2)
sns.scatterplot(data = telco, x = 'Eve_Charge', y = 'Eve_Mins',
                hue = 'Churn', alpha = 0.6)

plt.subplot(1,3,3)
sns.scatterplot(data = telco, x = 'Night_Charge', y = 'Night_Charge',
                hue = 'Churn', alpha = 0.6)

plt.show()
```

```python
dftotal = pd.DataFrame(telco['Churn'])
```

```python
dftotal['Total minutes'] = telco['Day_Mins'] + telco['Eve_Mins'] + telco['Night_Mins']
dftotal['Total calls'] = telco['Day_Calls'] + telco['Eve_Calls'] + telco['Night_Calls']
dftotal['Total charges'] = telco['Day_Charge'] + telco['Eve_Charge'] + telco['Night_Charge']
```

```python
dftotal
```

```python
plt.figure(figsize=(18,15))
plt.subplot(1,3,1)
sns.boxplot(data = dftotal, y = 'Total minutes',x='Churn', color = 'lightblue')
plt.ylabel('Total minutes', fontsize = 15)

plt.subplot(1,3,2)
sns.boxplot(data = dftotal, y = 'Total calls',x='Churn' ,color = 'lightgreen')
plt.ylabel('Total calls', fontsize = 15)

plt.subplot(1,3,3)
sns.boxplot(data = dftotal, y = 'Total charges',x='Churn', color = 'royalblue')
plt.ylabel('Total Charges', fontsize = 15)
```
